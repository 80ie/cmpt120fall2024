To link to an inline image/file in google Colab from Google drive, do the follow:

1. From Google Drive get the URL. Make sure it is shared as public and viewable. It will have something like this format: `https://drive.google.com/file/d/IMAGE_ID/view?usp=drive_link`. For example: `https://drive.google.com/file/d/11NKh26KF0L8bCQdj-UaBnrUw7LgRzKSq/view?usp=drive_link`.
2. Copy the IMAGE_ID from the above link into a link of this format: `https://drive.google.com/uc?id=IMAGE_ID`. For example, `https://drive.google.com/uc?id=11NKh26KF0L8bCQdj-UaBnrUw7LgRzKSq`.
