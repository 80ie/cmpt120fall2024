# Algorithms 2: Binary Search

## The Binary Search Algorithm

**Binary search** is a divide-and-conquer search algorithm that searches for a
target value on a sorted lists (i.e. lists whose values are in order from
smallest to biggest). Given a sorted list of values, it quickly finds the index
of a target value $x$ in the list, or returns -1 if $x$ is not found.

As we will see, binary search is *much* faster than linear search, especially
when searching large lists.


## A Binary Search Algorithm

Here's pseudocode for the binary search algorithm. It is essential that the list
be in sorted order from smallest to biggest:

1. Check if the middle element of the list is equal to $x$. If so, we're done.
2. If $x$ is less than the middle element, then $x$ must be in the left half of
   the list. So apply binary search to the left half.
3. If $x$ is greater than the middle element, then $x$ must be in the right half
   of the list. So apply binary search to the right half.

Each step of the algorithm reduces the size of the list to be searched by
*half*. This quickly narrows down the search to a small number of elements.

## Example Trace

Suppose we want to find 5 in the sorted list `[0, 2, 3, 4, 8, 9, 10]`. We will
use variables `lo` and `hi` to keep track of the current sub-list that is being
searched. Initially, we search the entire list:

```
looking for x = 5

[0, 2, 3, 4, 8, 9, 10]
 ^                  ^
 |                  |
lo                  hi
```

The middle element is 4. Since 5 is greater than 4, if 5 is in the list it must
be in the right half. So we adjust `lo`:

```
looking for x = 5

[0, 2, 3, 4, 8, 9, 10]
             ^     ^
             |     |
            lo    hi
```

The middle element of the current sub-list is 9. Since 5 is less than 9, if 5 is
in the list it must be in the left half. So we adjust `hi`:

```
looking for x = 5

[0, 2, 3, 4, 8, 9, 10]
             ^
             |
           lo, hi
```

Both `lo` and `hi` are pointing to the same element, and that element is not 5.
So we're done: 5 is not in the list.


## Binary Search Implementations

Here is a loop-based implementation of binary search:

````python
def binary_search_loop(x, lst):
    """Returns an index i such that list[i] == x.
    If x is not in lst, returns -1.
    lst must be in sorted order, from smallest to biggest.
    """
    lo = 0
    hi = len(lst) - 1
    while lo <= hi:
        mid = (lo + hi) // 2  # // is integer division
        if lst[mid] == x:
            return mid        # x found at location mid
        elif x < lst[mid]:
            hi = mid - 1
        else:                 # lst[mid] < x
            lo = mid + 1
    return -1                 # x not in lst

def binary_search_loop_test():
    """Test that binary search works correctly.
    assert expr does nothing if expr evaluates to True. If expr evaluates to
    False, then the program immediately crashes with an error message.
    """
    assert binary_search_loop(5, []             ) == -1
    assert binary_search_loop(1, [3]            ) == -1
    assert binary_search_loop(3, [3]            ) == 0
    assert binary_search_loop(5, [3]            ) == -1
    assert binary_search_loop(0, [1, 3]         ) == -1
    assert binary_search_loop(1, [1, 3]         ) == 0
    assert binary_search_loop(2, [1, 3]         ) == -1
    assert binary_search_loop(3, [1, 3]         ) == 1
    assert binary_search_loop(5, [5, 5, 5]      ) in [0, 1, 2]
    assert binary_search_loop(5, [1, 2, 5, 4, 7]) ==  2
    assert binary_search_loop(7, [1, 2, 5, 4, 7]) ==  4
    print('all binary_search_loop tests passed!')

binary_search_loop_test()
````

To help ensure our implementation is correct, we also wrote the test function
`binary_search_test`. If any of the test cases fail, then `assert` crashes the
program with a line number and error message.

Here is a recursive implementation of binary search:

````python
def binary_search_recursive_helper(x, lst, lo, hi):
    """Returns an index i such that lo <= i <= hi and list[i] == x.
    If x is not found in the range lo to hi, returns -1.
    lst must be in sorted order, from smallest to biggest.
    """
    if lo > hi:
        return -1  # x not in lst
    else:
        mid = (lo + hi) // 2  # // is integer division
        if lst[mid] == x:
            return mid
        elif x < lst[mid]:
            return binary_search_recursive_helper(x, lst, lo, mid - 1)
        else:
            return binary_search_recursive_helper(x, lst, mid + 1, hi)

def binary_search_recursive(x, lst):
    """Returns an index i such that list[i] == x.
    If x is not in lst, returns -1.
    lst must be in sorted order, from smallest to biggest.
    """
    return binary_search_recursive_helper(x, lst, 0, len(lst) - 1)

def binary_search_recursive_test():
    """Test that binary search works correctly.
    assert expr does nothing if expr evaluates to True. If expr evaluates to
    False, then the program immediately crashes with an error message.
    """
    assert binary_search_recursive(5, []             ) == -1
    assert binary_search_recursive(1, [3]            ) == -1
    assert binary_search_recursive(3, [3]            ) == 0
    assert binary_search_recursive(5, [3]            ) == -1
    assert binary_search_recursive(0, [1, 3]         ) == -1
    assert binary_search_recursive(1, [1, 3]         ) == 0
    assert binary_search_recursive(2, [1, 3]         ) == -1
    assert binary_search_recursive(3, [1, 3]         ) == 1
    assert binary_search_recursive(5, [5, 5, 5]      ) in [0, 1, 2]
    assert binary_search_recursive(5, [1, 2, 5, 4, 7]) ==  2
    assert binary_search_recursive(7, [1, 2, 5, 4, 7]) ==  4
    print('all binary_search_recursive tests passed!')

binary_search_recursive_test()

````

The recursive version uses a helper function that takes two additional
parameters,`lo` and `hi`. These parameters keep track of the current sub-list
that is being searched, and are adjusted as the search progresses.

## Performance of Linear Search and Binary Search

We've seen that, in the worst case, linear search does $n$ comparisons to search
a list of $n$ items. To search a list of $n$ (sorted) items, binary search does
$\log_2 n$ comparisons, which is much smaller.

This table gives you an idea of how much faster binary search is than linear
search:

|    # items searched <br> n      | Linear search comparisons <br> $n$      | Binary search comparisons <br>$\log_2 n$ |
|-----------|-------------------------|---------------------------|
|      10   |                      10 |                3.32       |
|     100   |                     100 |                6.64       |
|    1,000  |                   1,000 |                9.97       |
|   10,000  |                  10,000 |               13.29       |
|  100,000  |                 100,000 |               16.61       |
| 1,000,000 |               1,000,000 |               19.93       |

The last row shows that, in the worst case, linear search does a million
comparisons to search a list of size 1 million, while binary search does about
20 comparisons. Binary search is about 50,000 times faster!

**Again a warning!** Binary search requires that the list it is searching be in
sorted order, and so the table assumes that the list for binary search is
already sorted. If you must also sort the data, then linear search can be
faster!
